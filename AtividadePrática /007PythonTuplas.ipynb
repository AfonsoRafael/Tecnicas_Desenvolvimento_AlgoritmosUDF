Tuplas são imutaveis, dentro de () e separados por ,

```python
# tuplas
frutas = ("laranja", "maçã", "uva",)# tuple tem que terminar com , para evitar precedencia em python
letras = tuple("Python") # tuple para indicar tupla
numeros = tuple([1,2,3,4,5])# convertendo uma lista em tupla
pais = ("Brasil",)
print(frutas)
print(letras)
print(numeros)
print(pais)
```

```python
#acesso direto
frutas = ("laranja", "maçã", "uva",)
print(frutas[0])# igual na lista
print(frutas[1])
print(frutas[2])
print(frutas[-1])# acessa o ultimo
print(frutas[-2])
```

```python
#tuplas alinhadas
matriz = (
    (1,2,3),
    (4,5,6),
)
print(matriz[0])
print(matriz[1])
print(matriz[0][0])
print(matriz[0][-1])
```

```python
#fatiamento mesma ideia que em lista
tupla = tuple("Python")
print(tupla[2:])
print(tupla[1:3])
print(tupla[0:3:2])
```

```python
#iterar mesma ideia que a lista
frutas = ("laranja", "maçã", "uva",)
for fruta in frutas:
  print(fruta)
for indice, fruta in enumerate(frutas):
  print(f"{indice}:{fruta}")
```

Muito menos metodos porque não é alteravel

```python
# temos o count, index e o len
frutas = ("laranja", "maçã", "uva","maçã",)
print(frutas.count("maçã"))
print(frutas.index("maçã"))
print(len(frutas) )
```
