Curso Dio Listas

```python
#Exemplos de listas
frutas = ["laranja", "maça","uva"] #forma padrão
fruta = [] #Lista vazia
letras = list("python") # Construtor list, onde cada letra é um elemento de letras
numeros = list(range(10)) #Construtor retorna um elemento para cada valor retornado do range
carro = ['Ferrari', 'F8', 4200000, 2020,'portas',True] # Forma padrao

print(frutas)
print(letras)
print(numeros)
print(carro)
```

```python
# Acesso direto, toda lista começa com indice 0
frutas = ["laranja", "maça","uva"]
print(frutas[0])
print(frutas[2])
print(frutas[-1]) #tambem tem indices negativos
```

```python
#Listas alinhadas, estruturas bicondicionais, linhas e colunas
matriz = [
    [00,14,15],
    [14,13,12],
    [10,18,19]
]
print(matriz[0])#pega a linha da tabela
print(matriz[1])
print(matriz[2])
print(matriz[0][1])#pega o elemento da tabela
```

```python
# Fatiamento (Forma onde podemos passar um valor inicial e um valor final)
# Podemos informar as linhas que o cursor deve pular
lista = ["p","y","t","h","o","n"]
print(lista[2:])#tudo contando a partir do index 2
print(lista[:2])# antes de index 2, posição de parada não é retornada
print(lista[1:3])# 1 ate antes do 3
print(lista[0:]) # tudo, campo vazio significa 2
print(lista[0:3:2]) # da posição 0 até a posição de parada 3, pulando de 2 em 2
print(lista[::]) # tudo
print(lista[::-1]) # tudo, mas o inverso
```

```python
# Iterar listas
carros = ["gol","celta","palio"]
for carro in carros:
    print(carro)
```

```python
# Função enumereite, serve para saber o numero do indice que estamos percorrendo
carros = ["gol","celta","palio"]
for indice, carro in enumerate(carros):
    print(f"{indice}: {carro}")
```

```python
# Compreensão de listas
# Usado quando queremos criar uma lista nova com baso no que ja temos(filtro)
# Ou aplicar uma modificação(função) a todos os valores de uma lista existente
numeros = [1,30,21,2,9,65,34]
pares = []
# Versão 1 FILTRO
for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)# Adiciona novos valores a uma lista
print(pares)
# Versão 2 FILTRO
impares = [numero for numero in numeros if numero % 2 != 0]
# retorno gerado depois a iteração normal e por fim o if para o filtro
print(impares)
# Versão 1 MODIFICANDO VALORES
quadrado = []
for numero in numeros:
    quadrado.append(numero**2)
print(quadrado)
# Versão 2 MODIFICANDO VALORES
quadrado2 = [numero**2 for numero in numeros]
print(quadrado2)
```

METODOS E FUNÇÕES

```python
# append , adicionar elementos, adicionados no final da lista
lista = [] # por enquanto vazio
print(lista)
lista.append(1)
lista.append("Python")
lista.append([5,3,6])
print(lista)
```

```python
# clear , limpar a lista
lista = [1,"Python",[5,3,6]]
print(lista)
lista.clear()
print(lista)
```

```python
# copy, serve para criar uma copia da lista
lista = [1,"Python",[5,3,6]]
print(lista)
lista2 = lista.copy()
print(lista2)
print(id(lista2),id(lista))#Isso mostra que são objetos diferentes
lista2[0] = 6
print(lista2)
```

```python
# count, é usado para mostrar quantas vezes um elemento aparece dentro da lista
lista = [1,2,3,2,2,2,5,6,7,8,10,1,3,1,3]
print(lista.count(2))
print(lista.count(1))
print(lista.count(10))
```

```python
# extend, serve para juntar duas listas em uma
lista = [1,2,3]
print(lista)
lista.extend([4,5,6,1])# mesmo os repetidos
print(lista)


```

```python
# index, onde ocorre a primeira ocorrendo de um objeto
lista = [1,2,3,4,5,3,5,6,7,8]
print(lista.index(5))
print(lista.index(8))
print(lista.index(1))
```

```python
# pop , por padrão é uma pilha, primeiro a colocar ultimo a sair, ultimo a colocar primeiro a sair
lista = [1,2,3,4,5,3,5,6,7,8]
print(lista)
print(lista.pop())# ultimo a sair
print(lista)
print(lista.pop())# ultimo a sair
print(lista)
print(lista.pop())# ultimo a sair
print(lista)
print(lista.pop())# ultimo a sair
print(lista)
print(lista.pop(0))# podemos passar o primeiro indice tambem
print(lista)
```

```python
# remove, uma forma diferente de remover elementos
# em vez de passar o indice, passamos o elemento em si
lista = ["vasco","palmeiras","corinthians","gremio","atletico"]
print(lista)
lista.remove("corinthians")
print(lista)
```

```python
# reverse, serve para fazer o espelho da lista
lista = ["vasco","palmeiras","corinthians","gremio","atletico"]
print(lista)
lista.reverse()
print(lista)
```

```python
# sort serve para ordenar a lista
lista = ["vasco", "palmeiras", "corinthians", "gremio", "atletico"]

print(lista)               # lista original
lista.sort()               # ordena em ordem alfabética PADRÂO
print(lista)               # lista ordenada (A -> Z)
lista.sort(reverse=True)   # ordena em ordem reversa (Z -> A)
print(lista)               # lista invertida
lista.sort(key=lambda x: len(x))  # ordena pela quantidade de caracteres menor para o maior
print(lista)               # lista ordenada
lista.sort(key=lambda x: len(x), reverse=True)  # ordena pela quantidade de caracteres maior para o menor
print(lista)# se tiver o mesmo tamanho, o primeiro que estiver vem primeiro
```

```python
#len, serve para ver o tamanho da lista,
lista = ["vasco", "palmeiras", "corinthians", "gremio", "atletico"]
print(len(lista))
print(len(lista[1]))
```

```python
# sorted, tambem serve para ordenar, não altera a lista original
lista = ["vasco", "palmeiras", "corinthians", "gremio", "atletico"]

lista_ordenada = sorted(lista, key=lambda x: len(x))
print(lista_ordenada)  # ordem crescente pelo tamanho

lista_ordenada_reversa = sorted(lista, key=lambda x: len(x), reverse=True)
print(lista_ordenada_reversa)  # ordem decrescente
print(sorted(lista))
```
