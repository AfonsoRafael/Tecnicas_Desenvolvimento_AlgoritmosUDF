Dicionarios

É um conjunto de chave:valor não ordenados, são delititados por {} e contem lista de chabe:valor separada por virgula, onde a chave é uma instancia, imuttavel

```python
# Exemplo
pessoa = {"nome": "João", "idade": 25, "cidade": "São Paulo"}
print(pessoa)
pessoa = dict(nome="João", idade=25, cidade="São Paulo")# Mesma declaração so que com a classe dict
pessoa["telefone"] = "123456789" #Adicionando uma nova chave
print(pessoa)
print(pessoa["nome"]) #Acessando o valor da chave
pessoa["nome"] = "Afonso" # Alterando valor
print(pessoa)
```

```python
# Dicionarios alinhado, muito util com banco de dados
contatos = {
    "João": {"telefone": "1", "email": "2"},
    "Maria": {"telefone": "3", "email": "4"},
    "Pedro": {"telefone": "5", "email": "6", "extra" : {"teste": "A"}}
}
print(contatos)
print(contatos["João"])
print(contatos["João"]["telefone"])
extra = contatos["Pedro"]["extra"]["teste"]
print(extra)
```

```python
# Iterar dicionarios
contatos = {
    "João": {"telefone": "1", "email": "2"},
    "Maria": {"telefone": "3", "email": "4"},
    "Pedro": {"telefone": "5", "email": "6"}
}
for contato in contatos:
    print(contato)
    print(contatos[contato])
print("____________________________________________________________")
for chave in pessoa:
    print(chave, pessoa[chave])
print("____________________________________________________________")
for chave, valor in pessoa.items(): #Forma mais legivel
    print(chave, valor)
```

```python
# clear, apaga todos os valores
pessoa = {"nome": "João", "idade": 25, "cidade": "São Paulo"}
print(pessoa)
pessoa.clear()
print(pessoa)
```

```python
# copy, copia o dicionarioa
pessoa = {"nome": "João", "idade": 25, "cidade": "São Paulo"}
print(pessoa)
pessoa2 = pessoa.copy()
print(pessoa2)
pessoa2["nome"] = "Afonso"
print(pessoa)
print(pessoa2)
```

```python
# fromkeys, ele cria as chaves com valores vazios
dici = dict.fromkeys(["nome", "idade"]) # cria a chave sem valor associado
print(dici)
dicionario = dict.fromkeys(["nome", "idade"], "vazio") # fica com o campo padrão escrito vazio
print(dicionario)
# pode ser em dicionario existente ja ou criando um novo
```

```python
# get
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
#print(pessoa["chave"]) # isso gera um erro de excessão pois não existe essa chave no dicionario
print(pessoa.get("chave")) # excessão, serve para ver se a chave existe, se não existe retorna NONE, assim nãp da erro no codigo
print(pessoa.get("contato"))
print(pessoa.get("contato").get("nome"))
print(pessoa.get("contato" , {} )) # nesse caso como existe a chave retorna so o valor dela
print(pessoa.get("chave",{}))# mesma coisa da excessão, mas quando não encontra a chave ele retorna o valor que voce informar depois no caso {}
```

```python
# items ,retorna uma lista de tuplas
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa.items())
# muito util para usar o comando for para iterar os valores
```

```python
# keys retorna so as chaves do dicionario
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa.keys())
print(pessoa["contato"].keys())
# util para saber as chaves que o dicionario tem
```

```python
# pop remove uma chave do seu dicionario
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa["contato"].pop("nome", {}))#removendo uma chave e se ela não existir retorna o valor padrão
print(pessoa)
print(pessoa["contato"].pop("nome", "não encontrado"))#removendo uma chave e se ela não existir retorna o valor padrão
```

```python
#popitem remove chave tambem, mas não informamos qual é a chave e ele vai retirando os itens na sequencia
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa)
print(pessoa["contato"].popitem())
print(pessoa)
print(pessoa["contato"].popitem())
print(pessoa)
print(pessoa["contato"].popitem())
print(pessoa)
'''print(pessoa["contato"].popitem()) # não tem mais nada para remover e da erro de keyError
print(pessoa)'''
```

```python
# setdefault , se não exitir o valor ele adiciona, se existir ele informa o existente
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa)
print(pessoa['contato'].setdefault("nome", "Afonso")) # retorna joão por que ele existe
print(pessoa)
print(pessoa["contato"].setdefault("pai", "Afonso Silva")) # adiciona pai porque ainda não tinha
print(pessoa)
```

```python
#update, permite atualizar o dicionario
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"}}
print(pessoa)
pessoa.update({"contato": {"nome":"Afonso"}}) # Atualiza o dicionario tudo e mantem so o que foi declarado nele
print(pessoa)
pessoa.update({"contato": {"nome":"Afonso", "idade": 25, "cidade": "Brasilia"}})
print(pessoa)
pessoa.update({"contato1": {"nome": "João", "idade": 25, "cidade": "São Paulo"}}) # Cria outra chave com os elementos dentro
print(pessoa)
```

```python
# values faz retornar so os valores que estão amarados na chave
pessoa = {"contato" : {"nome": "João", "idade": 25, "cidade": "São Paulo"},
          "contato1": {"nome": "Afonso", "idade": 22, "cidade": "Brasilia"},
          "contado2": {"nome": "Carol", "idade": 20, "cidade": "Brasilia"}}
print(pessoa,"\n\n\n")
print(pessoa.values())
# muito util para iterar sobre os valores
```
