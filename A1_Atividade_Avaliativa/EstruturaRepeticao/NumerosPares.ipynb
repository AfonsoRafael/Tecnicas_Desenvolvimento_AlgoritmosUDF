```python
#Desenvolver um contador de 1 a 100 com for e outro com while, exibindo apenas números pares
print('Numeros pares com for')
for i in range (0, 101, 2):
  print(f'Numero: {i}')

print('--------------------------------------------')
print('Numeros pares com while')
cont = 1
while cont <= 100:
  if cont % 2 == 0:
    print(f'O numero {cont} é par')
  cont += 1
```
