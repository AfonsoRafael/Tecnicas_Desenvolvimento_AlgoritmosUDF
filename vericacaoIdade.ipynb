```python
# Sistema de verificação de idade para entrar em determinado filme
# Podera entrar maiores de 18 anos e maiores de 16 acompanhados de um maior de 18
cliente_nome = input("Qual o seu nome? ")
cliente_idade = int(input("Qual a sua idade? "))
if cliente_idade >= 18:
    print("Você pode entrar no filme, "+ cliente_nome + "!")
elif cliente_idade >= 16:
  acompanhante = input("Você tem acompanhante maior de idade(18)?S/N ")
  if acompanhante == "S" or "s":
    print("Você pode entrar no filme, "+ cliente_nome + "!")
  else:
    print("Você não pode entrar no filme, "+ cliente_nome + "! Pois precisa de um acompanhante")
else:
  print("Você não pode entrar no filme, "+ cliente_nome + "! Não tem idade o suficiente")
```
